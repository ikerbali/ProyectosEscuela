```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def pdf(alpha,beta):
    x=np.linspace(0,100,100)
    A=math.gamma(alpha)*(beta**alpha)
    p=(1/A)*(x**(alpha-1))*(np.exp(-x/beta))
    plt.plot(x,p,color="darkblue")
    plt.title("Gamma Probability Density Function")

for i in range(1,10,1):
    for j in range(1,10,1):
        plt.subplot(2,5,i)
        pdf(i,j)
```
