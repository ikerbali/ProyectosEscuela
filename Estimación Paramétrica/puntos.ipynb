```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

P1 = np.array([0, 0])
P2 = np.array([2 / np.sqrt(5), 4 / np.sqrt(5)])
P3 = np.array([np.sqrt(5), 0])

x_coords = [P1[0], P2[0], P3[0]]
y_coords = [P1[1], P2[1], P3[1]]

P4 = np.array([-12 / np.sqrt(5), 6 / np.sqrt(5)])
P5 = np.array([-7 / np.sqrt(5), 6/ np.sqrt(5)])
P6 = np.array([-9 / np.sqrt(5), 7 / np.sqrt(5)])
P7 = np.array([-2*np.sqrt(5), 2*np.sqrt(5)])


x_coords_1 = [P4[0], P5[0], P6[0],P7[0]]
y_coords_1= [P4[1], P5[1], P6[1],P7[1]]

plt.scatter(x_coords_1, y_coords_1, color="blue",label="Puntos de Equipo 1")
plt.scatter(x_coords, y_coords, color="red",label="Puntos de Equipo 2")


X=np.linspace(-10,10,100)
a=7/(2*np.sqrt(5))
y=7/6*(X+a)+3/np.sqrt(5)
plt.plot(X,y,color="purple",label="curva del lobo")


# curva de jp


X_1=np.linspace(-10,10,100)
a_1=np.sqrt(5)/2
y_1=-9/2*(X+a_1)+np.sqrt(5)
plt.plot(X_1,y_1,color="green",label="curva de jp")


# curva de cons


X_2=np.linspace(-10,10,100)
a_2=3/np.sqrt(5)
y_2=2*(X_2+a_2)+4/np.sqrt(5) 
plt.plot(X_2,y_2,color="blue",label="curva de cons")
plt.axhline(2)
plt.legend()
plt.show()
```
